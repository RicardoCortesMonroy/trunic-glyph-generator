## Generate csv file that can be imported to Anki

The following is a script to generate a csv file that can be imported to Anki as deck. The deck contains to the top `n` most freqently used English words, with their transcriptions into Trunic on the front side and the English spelling on the back.

```python
import pandas as pd
from trunic_generator import english_to_trunic

# Number of words to add to deck
# Deck will contain the n most frequent english words (up until 333,334)
n = 100

df = pd.read_csv("Data/unigram_freq.csv")
df = df.head(n)

words_concatenated = ';'.join(df["word"])
print(words_concatenated)

unicode_list = english_to_trunic(words_concatenated)
tunic_unicode_concatenated = ''.join([c if not isinstance(c,int) else chr(c) for c in unicode_list])

deck = pd.DataFrame()
deck["front"] = tunic_unicode_concatenated.split(';')
deck["back"] = df["word"]

display(deck)
out_filename = f"Data/english-trunic-{n}.csv"
deck.to_csv(out_filename, index=False, header=False)
```
